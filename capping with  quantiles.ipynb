#### Capping 
is a technique used in feature engineering to handle outliers in a dataset. Outliers are data points that lie far from the majority of the other data points in a dataset, and they can have a significant impact on the analysis and modeling of the data. Capping involves setting a limit or cap on the values of a feature, such that any values that exceed this limit are replaced with the limit value.

For example, if we have a dataset of housing prices, and there are a few extreme values that are much higher than the rest of the values, we can use capping to limit the impact of these outliers on the analysis. We might choose to cap the values of the feature at the 99th percentile, such that any values that exceed this percentile are replaced with the value at the 99th percentile. This can help to make the dataset more robust to outliers and ensure that the analysis is not overly influenced by these extreme values.

Capping can be performed in different ways, such as using a fixed value, a percentile value, or a multiple of the standard deviation. The choice of capping method will depend on the characteristics of the data and the specific requirements of the analysis. It is important to note that capping can have both positive and negative effects on the analysis, and should be used judiciously and with careful consideration of the potential impacts on the results.

### with upper and lower quartile

import pandas as pd
import numpy as np

data = pd.read_csv('example.csv')  # load the dataset

q1 = data['Income'].quantile(0.25)  # calculate the 25th percentile
q3 = data['Income'].quantile(0.75)  # calculate the 75th percentile
iqr = q3 - q1  # calculate the interquartile range

lower_cap = q1 - 1.5 * iqr  # calculate the lower cap value
upper_cap = q3 + 1.5 * iqr  # calculate the upper cap value

data.loc[data['Income'] < lower_cap, 'Income'] = lower_cap  # cap income values below the lower threshold
data.loc[data['Income'] > upper_cap, 'Income'] = upper_cap  # cap income values above the upper threshold
