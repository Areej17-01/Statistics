## CDF

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
x = np.random.normal(size=1000)

# Calculate CDF
hist, bin_edges = np.histogram(x, bins=100, density=True)
cdf = np.cumsum(hist * np.diff(bin_edges))

# Plot CDF
plt.plot(bin_edges[1:], cdf)
plt.title("CDF of a Normal Distribution")
plt.xlabel("x")
plt.ylabel("Cumulative Probability")
plt.show()
```

## PDF

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
x = np.linspace(-5, 5, num=1000)

# Calculate PDF
pdf = np.exp(-0.5*x**2) / np.sqrt(2*np.pi)

# Plot PDF
plt.plot(x, pdf)
plt.title("PDF of a Normal Distribution")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.show()
```

### PMF

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
x = np.arange(1, 7)
p = np.array([1/6, 1/6, 1/6, 1/6, 1/6, 1/6])

# Plot PMF
plt.stem(x, p, use_line_collection=True)
plt.title("PMF of a Fair Die")
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.show()
```

### THEORY:

In probability theory, the probability mass function (PMF) and the cumulative distribution function (CDF) are two fundamental concepts used to describe the probability distribution of a random variable.

The PMF is a function that gives the probability of each possible value that a discrete random variable can take. For example, consider a fair six-sided die. The PMF for this die would be:

P(X = 1) = 1/6, P(X = 2) = 1/6, P(X = 3) = 1/6,
P(X = 4) = 1/6, P(X = 5) = 1/6, P(X = 6) = 1/6

This means that the probability of rolling a 1 is 1/6, the probability of rolling a 2 is 1/6, and so on.

The CDF, on the other hand, gives the probability that a random variable takes a value less than or equal to a given value. The CDF is defined for both discrete and continuous random variables. For example, the CDF for the same fair six-sided die would be:

F(x) = P(X ≤ x)
= 0 if x < 1
= 1/6 if 1 ≤ x < 2
= 1/3 if 2 ≤ x < 3
= 1/2 if 3 ≤ x < 4
= 2/3 if 4 ≤ x < 5
= 5/6 if 5 ≤ x < 6
= 1 if x ≥ 6

This means that the probability of rolling a number less than or equal to 1 is 0, the probability of rolling a number less than or equal to 2 is 1/6, the probability of rolling a number less than or equal to 3 is 1/3, and so on.

To illustrate the concept of PMF and CDF with an example related to the binomial distribution, consider a scenario where a basketball player has a free-throw percentage of 80%. We can model the number of successful free-throws in 10 attempts as a binomial distribution with parameters n = 10 and p = 0.8.

The PMF for this binomial distribution would give the probability of observing exactly k successful free-throws in 10 attempts, for all possible values of k between 0 and 10. For example, the PMF for k = 3 would give the probability of observing exactly 3 successful free-throws in 10 attempts

The CDF for this binomial distribution would give the probability of observing k or fewer successful free-throws in 10 attempts, for all possible values of k between 0 and 10. For example, the CDF for k = 3 would give the probability of observing 3 or fewer successful free-throws in 10 attempts.
