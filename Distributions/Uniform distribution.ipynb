```python
from scipy.stats import uniform
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import randint
```

# Uniform pdf

The probability density function (pdf) of uniform distribution is defined as:

        f(x) = 1/(b-a)     a<=x<=b,
               0           otherwise

Uniform Distribution
Where, a and b are lower and upper boundaries of output interval respectively.

An uniform distribution has mean (a+b)/2 and variance (b-a)2/12.

```python
#creating an array of values ranging from -5 to 10 with 0.2 difference
x = np.arange(-5,10,0.2)

y = uniform.pdf(x, 0, 5)

plt.plot(x, y) 
plt.show()
```

# Unifrom cdf

The cumulative distribution function (cdf) evaluated at x, is the probability that the random variable (X) will take a value less than or equal to x. The cdf of uniform distribution is defined as:

        F(x) = P(X<=x) = 0                x<a
                         (x-a)/(b-a)      a<=x<=b
                         1                x>b

```python
x = np.arange(-5,10,0.2)

y = uniform.cdf(x, 0, 5)

plt.plot(x, y) 
plt.show()
```

```python
a=1  #  rolled once
b=6  #  six sides

x = np.arange(a, b+1)
print(x)

dis_uniform_dist = randint(a, b+1)
dis_uniform_pmf = dis_uniform_dist.pmf(x)
print(dis_uniform_pmf)

plt.plot(x, dis_uniform_pmf,'bo', ms=8)
plt.vlines(x, 0, dis_uniform_pmf, colors='r')
plt.xlabel('Sides of the Dice')
plt.ylabel('Probability')
plt.show()
```
