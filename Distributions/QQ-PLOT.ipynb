A QQ plot (quantile-quantile plot) is a graphical tool used to compare the distribution of a sample to a known probability distribution. It is often used in statistics to assess whether a sample of data comes from a population with a specific distribution, such as the normal distribution.

The basic idea behind a QQ plot is to plot the quantiles of the sample data against the corresponding quantiles of the theoretical distribution. If the sample data follows the theoretical distribution, the plot will follow a straight line. Deviations from the straight line indicate departures from the theoretical distribution.

To construct a QQ plot, follow these steps:

Sort the data in ascending order.
Calculate the percentiles of the data using the formula (i - 0.5) / n, where i is the rank of the data point (i.e., its position in the sorted list) and n is the total number of data points.
Calculate the corresponding quantiles of the theoretical distribution. This can be done using a statistical software package or by hand using tables of the cumulative distribution function (CDF) for the distribution in question.
Plot the percentiles of the data against the corresponding quantiles of the theoretical distribution.
If the data follows the theoretical distribution, the points on the QQ plot will fall on a straight line. If the data deviates from the theoretical distribution, the points will deviate from the line.

In summary, a QQ plot is a useful graphical tool for assessing whether a sample of data comes from a population with a specific distribution. By comparing the quantiles of the sample data to the corresponding quantiles of the theoretical distribution, we can detect departures from the expected distribution.




```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats
```

## Example Plots

```python
mu, sigma = 5, 1 # mean and standard deviation
s = np.random.normal(mu, sigma, 1000)
nflx=s
nflx = np.diff(np.log(nflx[nflx>0]))
fig, ax = plt.subplots(figsize=(4, 4))
values=stats.probplot(nflx, plot=ax)
```

##### doesnot shows too much extreme values ,which tells that it do  follows normal distribution as it is following the linear line.

```python
marks = pd.Series([10,50,28,50,50,58,60,59,60,57,60,60,71,68,65,67,100,101,100,101,100,101])
nflx=marks
nflx = np.diff(np.log(nflx[nflx>0]))
fig, ax = plt.subplots(figsize=(4, 4))
stats.probplot(nflx, plot=ax)
```
