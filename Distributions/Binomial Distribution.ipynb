```python
import numpy as np
import matplotlib.pyplot as plt

# Set parameters
n = 1000  # number of trials
p = 0.5  # probability of success in each trial

# Generate data
x = np.random.binomial(n, p, size=10000)

# Plot histogram
plt.hist(x, bins=20)
plt.title("Histogram of a Binomial Distribution (n={}, p={})".format(n, p))
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```

### THEORY

The binomial distribution is a probability distribution that describes the number of successes in a fixed number of independent trials, given a constant probability of success for each trial. The binomial distribution is characterized by two parameters: n, the number of trials, and p, the probability of success in each trial. The probability of exactly x successes in n trials is given by the probability mass function (PMF) of the binomial distribution.

To identify that the distribution is binomial, we can look for several characteristics:

The distribution is discrete and takes on integer values.

The distribution is bounded by 0 and n.

The distribution has a fixed number of trials n and a fixed probability of success in each trial p.

The distribution has a "bell curve" shape with its peak at or near the expected number of successes np.

In the plot generated by the code above, we can see that the histogram has a "bell curve" shape with its peak at the expected value of 500 (i.e., np=1000*0.5=500), which is a characteristic of the binomial distribution. The distribution is also bounded by 0 and 1000 and takes on only integer values.

### is rolling a dice binomial?
Rolling a single fair dice is not an example of a binomial distribution. The binomial distribution is used to model the number of successes in a fixed number of independent Bernoulli trials, where each trial has a constant probability of success.

In the case of rolling a dice, we only have one trial with six possible outcomes, and the probability of each outcome is not constant, but rather uniform. Therefore, we cannot use the binomial distribution to model the outcome of rolling a dice.

Instead, the outcome of rolling a dice can be modeled using a discrete uniform distribution, where each of the six possible outcomes has an equal probability of 1/6.
