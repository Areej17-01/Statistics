```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats
```

```python
import random

# Generate a random float number between 0 and 1 with uniform distribution
x = random.uniform(0, 1)

# Generate 10 random float numbers between 0 and 1 with uniform distribution
n = 10000
x_list = [random.uniform(0, 1) for i in range(n)]
```

```python
x_list[:5]
```

### PLOT

```python
marks=pd.Series(x_list)
marks.plot.kde(label = "marks")

plt.axvline(marks.mean(), color = "green",label="Mean")
plt.axvline(marks.median(), color = "red",label="Median")
plt.axvline(marks.mode().mean(), color = "black",label="Mode")

plt.legend()
plt.show()
```

### BY SCIPY

###    QQ-PLOT

```python
nflx=marks
nflx = np.diff(np.log(nflx[nflx>0]))
fig, ax = plt.subplots(figsize=(4, 4))
values=stats.probplot(nflx, plot=ax)
```
