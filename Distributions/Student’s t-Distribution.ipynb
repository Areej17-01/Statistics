```python
import numpy as np

# degrees of freedom (sample size - 1)
df = 10
# generate random numbers from t-distribution
x = np.random.standard_t(df, size=90)

x[:10]
```

```python
# print the mean and standard deviation of the generated numbers
print("Mean:", np.mean(x))
print("Standard deviation:", np.std(x))
```

```python
t_dist=pd.Series(x)
```

```python
t_dist.plot.kde(label = "t_dist")
plt.axvline(t_dist.mean(), color = "green",label="Mean")
plt.axvline(t_dist.median(), color = "red",label="Median")
plt.axvline(t_dist.mode().mean(), color = "black",label="Mode")

plt.legend()
plt.show()
```

## EXPLAINATION

he Student's t-distribution, also known as the t-distribution, is a probability distribution that is used to estimate population parameters when the sample size is small and the population standard deviation is unknown. It is a bell-shaped distribution that is similar to the normal distribution, but has heavier tails, meaning it has more probability in the tails than the normal distribution.

The t-distribution is often used in hypothesis testing and confidence interval calculations, where the sample size is small and the population standard deviation is unknown. For example, suppose a researcher wants to estimate the average height of a population of students, but can only measure the height of a small sample of students. If the sample size is small and the population standard deviation is unknown, the t-distribution can be used to calculate the confidence interval for the population mean.

An example of the t-distribution in action could be a study where a researcher collects data on the heights of 10 randomly selected college students, and wants to estimate the population mean height with 95% confidence. The sample mean is 170cm, and the sample standard deviation is 5cm. Using the t-distribution with 9 degrees of freedom (since there are 10 students in the sample), the researcher can calculate the confidence interval for the population mean height as (163.28cm, 176.72cm). This means that the researcher is 95% confident that the true population mean height lies between 163.28cm and 176.72cm.

### DEGREE OF FREEDOM

The degree of freedom (df) is a statistical concept that refers to the number of values in a calculation that are free to vary. It is typically used in the context of hypothesis testing, where the degree of freedom is used to determine the appropriate statistical distribution to use in a test.

In general, the degree of freedom is equal to the sample size minus the number of parameters estimated from the sample. For example, if we have a sample of size 10 and we are estimating the mean, then we have one parameter estimated from the sample, and the degree of freedom would be 10 - 1 = 9.

In the case of the Student's t-distribution, the degree of freedom is used to determine the shape of the distribution. As the degree of freedom increases, the t-distribution becomes more similar to the normal distribution. When the degree of freedom is large (e.g. greater than 30), the t-distribution is practically indistinguishable from the normal distribution.

In summary, the degree of freedom is a parameter that determines the number of independent observations in a sample and is used to calculate statistical properties of the sample, such as the standard deviation, confidence intervals, and the shape of the distribution.
